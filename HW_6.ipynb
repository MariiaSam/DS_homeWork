```python
# Ð“Ñ–Ð¿Ð¾Ñ‚ÐµÐ·Ð° (Ñ„ÑƒÐ½ÐºÑ†Ñ–Ñ Ð¿ÐµÑ€ÐµÐ´Ð±Ð°Ñ‡ÐµÐ½Ð½Ñ)
def hypothesis(X, w):
    return np.dot(X, w)
```

```python
# Ð¤ÑƒÐ½ÐºÑ†Ñ–Ñ Ð´Ð»Ñ Ð¾Ð±Ñ‡Ð¸ÑÐ»ÐµÐ½Ð½Ñ Ð²Ñ‚Ñ€Ð°Ñ‚ (MSE)
def compute_cost(X, y, w):
    m = len(y)
    error = hypothesis(X, w) - y
    return np.sum(error ** 2) / (2 * m)
```

```python
# ÐžÐ´Ð¸Ð½ ÐºÑ€Ð¾Ðº Ð³Ñ€Ð°Ð´Ñ–Ñ”Ð½Ñ‚Ð½Ð¾Ð³Ð¾ ÑÐ¿ÑƒÑÐºÑƒ
def gradient_step(X, y, w, alpha=0.00000001):
    m = len(y)
    error = hypothesis(X, w) - y
    gradient = (1 / m) * np.dot(X.T, error)
    w = w - alpha * gradient
    return w
```

```python
# Ð“Ñ€Ð°Ð´Ñ–Ñ”Ð½Ñ‚Ð½Ð¸Ð¹ ÑÐ¿ÑƒÑÐº
def gradient_descent(X, y, w, alpha, num_iters):
    cost_history = []

    for i in range(num_iters):
        w = gradient_step(X, y, w, alpha)
        cost = compute_cost(X, y, w)
        cost_history.append(cost)

    return w, cost_history
```

```python
# ÐÐ½Ð°Ð»Ñ–Ñ‚Ð¸Ñ‡Ð½Ðµ Ñ€Ñ–ÑˆÐµÐ½Ð½Ñ
def analytical_solution(X, y):
    X_transpose = np.transpose(X)
    w = np.linalg.inv(X_transpose.dot(X)).dot(X_transpose).dot(y)
    return w
```

```python
# ÐžÐ±Ñ‡Ð¸ÑÐ»ÐµÐ½Ð½Ñ ÑÐµÑ€ÐµÐ´Ð½ÑŒÐ¾ÐºÐ²Ð°Ð´Ñ€Ð°Ñ‚Ð¸Ñ‡Ð½Ð¾Ñ— Ð¿Ð¾Ð¼Ð¸Ð»ÐºÐ¸ (MSE)
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
# ÐžÑÐ½Ð¾Ð²Ð½Ð¸Ð¹ ÐºÐ¾Ð´
X = df[['area', 'bathrooms', 'bedrooms']].values
y = df['price'].values

# Ð”Ð¾Ð´Ð°Ñ”Ð¼Ð¾ ÑÑ‚Ð¾Ð²Ð¿ÐµÑ†ÑŒ Ð· 1 Ð´Ð»Ñ Ð²Ñ–Ð»ÑŒÐ½Ð¾Ð³Ð¾ Ñ‡Ð»ÐµÐ½Ð° (Ñ–Ð½Ñ‚ÐµÑ€ÑÐµÐ¿Ñ‚)
X = np.c_[np.ones(X.shape[0]), X]
```

```python
# 1. Ð“Ñ€Ð°Ð´Ñ–Ñ”Ð½Ñ‚Ð½Ð¸Ð¹ ÑÐ¿ÑƒÑÐº
initial_w = np.zeros(X.shape[1])
alpha = 0.00000001
num_iters = 10000
w_gd, cost_history = gradient_descent(X, y, initial_w, alpha, num_iters)
print(f'Ð’Ð°Ð³Ð¸ Ð· Ð³Ñ€Ð°Ð´Ñ–Ñ”Ð½Ñ‚Ð½Ð¾Ð³Ð¾ ÑÐ¿ÑƒÑÐºÑƒ: {w_gd}')

# 2. ÐÐ½Ð°Ð»Ñ–Ñ‚Ð¸Ñ‡Ð½Ðµ Ñ€Ñ–ÑˆÐµÐ½Ð½Ñ
w_analytical = analytical_solution(X, y)
print(f'Ð’Ð°Ð³Ð¸ Ð· Ð°Ð½Ð°Ð»Ñ–Ñ‚Ð¸Ñ‡Ð½Ð¾Ð³Ð¾ Ñ€Ñ–ÑˆÐµÐ½Ð½Ñ: {w_analytical}')
```

```python
# 3. Scikit-learn
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X[:, 1:], y)  # ÐŸÑ€Ð¾Ð¿ÑƒÑÐºÐ°Ñ”Ð¼Ð¾ Ð¿ÐµÑ€ÑˆÐ¸Ð¹ ÑÑ‚Ð¾Ð²Ð¿ÐµÑ†ÑŒ (Ñ–Ð½Ñ‚ÐµÑ€ÑÐµÐ¿Ñ‚)
w_sklearn = np.concatenate([[model.intercept_], model.coef_])
print(f'Ð’Ð°Ð³Ð¸ Ð· scikit-learn: {w_sklearn}')
```

```python
# ÐŸÑ€Ð¾Ð³Ð½Ð¾Ð·Ð¸ Ð´Ð»Ñ ÐºÐ¾Ð¶Ð½Ð¾Ð³Ð¾ Ð¼ÐµÑ‚Ð¾Ð´Ñƒ
y_pred_gd = hypothesis(X, w_gd)  # Ð“Ñ€Ð°Ð´Ñ–Ñ”Ð½Ñ‚Ð½Ð¸Ð¹ ÑÐ¿ÑƒÑÐº
y_pred_analytical = hypothesis(X, w_analytical)  # ÐÐ½Ð°Ð»Ñ–Ñ‚Ð¸Ñ‡Ð½Ðµ Ñ€Ñ–ÑˆÐµÐ½Ð½Ñ
y_pred_sklearn = hypothesis(X, w_sklearn)  # Scikit-learn
```

```python
# Ð’Ð¸Ð²Ð¾Ð´Ð¸Ð¼Ð¾ MSE Ð´Ð»Ñ ÐºÐ¾Ð¶Ð½Ð¾Ð³Ð¾ Ð¼ÐµÑ‚Ð¾Ð´Ñƒ
mse_gd = mse(y, y_pred_gd)
mse_analytical = mse(y, y_pred_analytical)
mse_sklearn = mse(y, y_pred_sklearn)

print(f'MSE Ð· Ð³Ñ€Ð°Ð´Ñ–Ñ”Ð½Ñ‚Ð½Ð¾Ð³Ð¾ ÑÐ¿ÑƒÑÐºÑƒ: {mse_gd}')
print(f'MSE Ð· Ð°Ð½Ð°Ð»Ñ–Ñ‚Ð¸Ñ‡Ð½Ð¾Ð³Ð¾ Ñ€Ñ–ÑˆÐµÐ½Ð½Ñ: {mse_analytical}')
print(f'MSE Ð· scikit-learn: {mse_sklearn}')
```
