```python
# Гіпотеза (функція передбачення)
def hypothesis(X, w):
    return np.dot(X, w)
```

```python
# Функція для обчислення втрат (MSE)
def compute_cost(X, y, w):
    m = len(y)
    error = hypothesis(X, w) - y
    return np.sum(error ** 2) / (2 * m)
```

```python
# Один крок градієнтного спуску
def gradient_step(X, y, w, alpha=0.00000001):
    m = len(y)
    error = hypothesis(X, w) - y
    gradient = (1 / m) * np.dot(X.T, error)
    w = w - alpha * gradient
    return w
```

```python
# Градієнтний спуск
def gradient_descent(X, y, w, alpha, num_iters):
    cost_history = []

    for i in range(num_iters):
        w = gradient_step(X, y, w, alpha)
        cost = compute_cost(X, y, w)
        cost_history.append(cost)

    return w, cost_history
```

```python
# Аналітичне рішення
def analytical_solution(X, y):
    X_transpose = np.transpose(X)
    w = np.linalg.inv(X_transpose.dot(X)).dot(X_transpose).dot(y)
    return w
```

```python
# Обчислення середньоквадратичної помилки (MSE)
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
# Основний код
X = df[['area', 'bathrooms', 'bedrooms']].values
y = df['price'].values

# Додаємо стовпець з 1 для вільного члена (інтерсепт)
X = np.c_[np.ones(X.shape[0]), X]
```

```python
# 1. Градієнтний спуск
initial_w = np.zeros(X.shape[1])
alpha = 0.00000001
num_iters = 10000
w_gd, cost_history = gradient_descent(X, y, initial_w, alpha, num_iters)
print(f'Ваги з градієнтного спуску: {w_gd}')

# 2. Аналітичне рішення
w_analytical = analytical_solution(X, y)
print(f'Ваги з аналітичного рішення: {w_analytical}')
```

```python
# 3. Scikit-learn
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X[:, 1:], y)  # Пропускаємо перший стовпець (інтерсепт)
w_sklearn = np.concatenate([[model.intercept_], model.coef_])
print(f'Ваги з scikit-learn: {w_sklearn}')
```

```python
# Прогнози для кожного методу
y_pred_gd = hypothesis(X, w_gd)  # Градієнтний спуск
y_pred_analytical = hypothesis(X, w_analytical)  # Аналітичне рішення
y_pred_sklearn = hypothesis(X, w_sklearn)  # Scikit-learn
```

```python
# Виводимо MSE для кожного методу
mse_gd = mse(y, y_pred_gd)
mse_analytical = mse(y, y_pred_analytical)
mse_sklearn = mse(y, y_pred_sklearn)

print(f'MSE з градієнтного спуску: {mse_gd}')
print(f'MSE з аналітичного рішення: {mse_analytical}')
print(f'MSE з scikit-learn: {mse_sklearn}')
```

значення ваг та середньоквадратичної помилки (MSE) між градієнтним спуском, аналітичним рішенням і Scikit-learn є подібними, це означає, що всі методи дали аналогічні результати
